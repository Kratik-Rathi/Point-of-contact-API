```python
import pandas as pd
import random
```

```python
df = pd.read_csv(r"D:\MS\Indiana University Bloomington\Luddy Hackathon 2024\data.csv")
```

```python
df
```

```python
technical_titles = [
    'Software Engineer', 'Backend Developer', 'Frontend Developer', 'DevOps Engineer', 'Cloud Architect',
    'Data Scientist', 'Machine Learning Engineer', 'Database Administrator', 'Full Stack Developer', 'Mobile App Developer',
    'Cybersecurity Analyst', 'QA Engineer', 'Product Manager', 'Systems Analyst', 'IT Support Specialist',
    'Network Engineer', 'UX/UI Designer', 'Technical Writer', 'IT Project Manager', 'AI Researcher'
]

df['Title'] = [random.choice(technical_titles) for _ in range(len(df))]
```

```python
products = [
    'SecurityScanAPI', 'PaymentGateway', 'UserAuthSystem', 'DataPipeline', 'PaymentSystem',
    'DataStorageService', 'CloudBackup', 'AnalyticsDashboard', 'UserManagement', 'ContentDeliveryAPI',
    'CustomerSupportChat', 'DocumentGenerator', 'NotificationService', 'APIIntegrationTool', 'SearchEngine',
    'UserActivityTracker', 'MarketingAutomation', 'ECommerceAPI', 'InvoiceSystem', 'ProductRecommendationEngine'
]

repositories = [
    'SecurityRepo', 'PaymentRepo', 'AuthRepo', 'DataRepo', 'PaymentSystemRepo',
    'BackupRepo', 'AnalyticsRepo', 'UserRepo', 'ContentRepo', 'DeliveryRepo',
    'SupportRepo', 'DocGenRepo', 'NotificationRepo', 'IntegrationRepo', 'SearchRepo',
    'ActivityRepo', 'MarketingRepo', 'EcommerceRepo', 'InvoiceRepo', 'RecommendationRepo'
]

df['ProductAssigned'] = [random.choice(products) for _ in range(len(df))]
df['RepositoryAssigned'] = [random.choice(repositories) for _ in range(len(df))]

print(df[['EmpID', 'FirstName', 'LastName', 'Location', 'ProductAssigned', 'RepositoryAssigned']].head())
```

```python
df.rename(columns={
    'ADEmail': 'EmailID',
    'ProductAssigned': 'ProductName',
    'RepositoryAssigned': 'RepositoryName'
}, inplace=True)

print(df.columns)

export_path = r"D:\MS\Indiana University Bloomington\Luddy Hackathon 2024\processed_data.csv"
df.to_csv(export_path, index=False)
```
